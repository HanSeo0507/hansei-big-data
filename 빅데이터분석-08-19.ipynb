```python
# 데이터의 시각화-멧플롯립(Matplotlib)
# 멧플롯립은 파이썬의 대표적인 과학 계산용 그래프 라이브러리
# 탐색적 데이터 분석과 데이터 결과의 시각화를 위해 자주 사용됨.
# 그래프를 그릴 때는 pyplot 부속 라이브러리를 사용하고 별칭은 관례적으로 plt

import pandas as pd
import matplotlib.pyplot as plt

# 선 그래프 - plot() 함수
# y축에 해당하는 데이터는 반드시 입력해야함.


plt.plot([10, 20, 30, 40])
# x축의 데이터는 자동으로 [0,1,2,3]으로 입력
plt.show()
# 그래프를 보여주기 위한 show() 함수
```

```python
# 선 그래프 - plot(x축, y축)
plt.title("plotting")
plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

```python
# 범례표시 - 그래프가 의미하는 바를 구별하기 위한 방법
# 법례는 보통 두 개 이상의 데이터를 표시할 때 사용합니다.
# plot() 함수에 label이라는 속성이 레이블 값으로 원하는 문자열을 넣어주고 
# 그래프를 그리기 전에 legend() 함수를 실행시키면 레이블 값이 범례로 나타남.
# 그래프의 선 모양응ㄹ 다양한 혀애로 바꾸고 싶을 때는 linestyle 속성에 원하는 선 모양을
# 지정하면 됩니다. 이 때 linestyle 속성 대신 ls라고 작성할 수도 있습니다

plt.title("legend")
plt.plot([10, 20, 30, 40], color="skyblue", linestyle="--", label="asc") # 증가를 의미하는 asc 범례
plt.plot([40, 30, 20, 10], "pink", ls=":", label="desc") # 감소를 의미하는 desc 범례
plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기
# plot() 함수에 marker 속성을 설정하면 선 형태가 아닌 점 형태로 그래프를 그릴 수 있음.
# 이때 색상과 마커 모양을 한 번에 설정할 수도 있음
# 마커 모양으로는 점, 선, 동그라미, 삼각형, 별모양 등 여러가지가 있음.
# 예를 들어 다음 코드에서 'r.'의 마침표는 (.)는 점 모양의 마커,
# g^의 ^는 삼각형 모양의 마커를 의미함

plt.title("marker") # 제목설정
plt.plot([10, 20, 30, 40], "r.", label="circle") # 빨간색 원형 마커 그래프
plt.plot([40, 30, 20, 10], "g^", label="triangle up") #
plt.legend()
plt.show()
```

```python
# numpy 라이브러리를 활용하면 다양한 함수들의 그래프를 표현할 수 있다.
# 또한 여러 개의 plot() 함수를 사용하면 한 번에 여러 개의 그래프를 그릴 수 있다.

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 2 * np.pi, 0.01)

# 2개의 plot 함수를 사용하여 2개의 그래프 그리기
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```
